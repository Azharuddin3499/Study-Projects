**Q: How many countries does the dataframe contain?**

```python
num_countries = countries_df.shape[0]
print("There are {} countries in countries_df".format(num_countries))
```

**Q: Retrieve a list of continents from the dataframe?**

**Q: What is the total population of all the countries listed in this dataset?**

```python
Total_population = countries_df.population.sum()
print("Total Populaton of all countries in contries_df is {} ".format(int(Total_population)))
```

**Q:What is the overall life expectancy across in the world?**

```python
overall_life_expectancy = (countries_df['population'] * countries_df['life_expectancy']).sum() / countries_df['population'].sum()
print("The overall life expectancy across the world is: {:.2f} years".format(overall_life_expectancy))
```

**Q: Create a dataframe containing 10 countries with the highest population.**

**Q: Add a new column in `countries_df` to record the overall GDP per country (product of population & per capita GDP).**

**Q: Create a dataframe containing 10 countries with the lowest GDP per capita, among the countries with population greater than 100 million.**

**Q: Count the number of countries for which the `total_tests` data is missing.**

```python
total_tests_missing = covid_data_df['total_tests'].isna().sum()
print("The data for total tests is missing for {} countries.".format(int(total_tests_missing)))
```

**Q: Merge countries_df with covid_data_df on the location column.**

```python
merge_df = countries_df.merge(covid_data_df ,on="location")
merge_df
```

**Q: Add columns `tests_per_million`, `cases_per_million` and `deaths_per_million` into `combined_df`.**

```python
merge_df['tests_per_million'] = merge_df['total_tests'] * 1e6 / merge_df['population']
merge_df['cases_per_million'] = merge_df['total_cases'] * 1e6 / merge_df['population']
merge_df['deaths_per_million'] = merge_df['total_deaths'] * 1e6 / merge_df['population']
```

**Q: Create a dataframe with 10 countires that have highest number of tests per million people.**

```python
highest_tests = merge_df.sort_values(by='tests_per_million', ascending=False).head(10)
highest_tests_df = pd.DataFrame(highest_tests).reset_index(drop =True)
highest_tests_df =highest_tests_df[[
                    'location',
                    'continent', 
                    'population', 
                    'cases_per_million',
                    'hospital_beds_per_thousand',
                    'gdp_per_capita'
                    ]]
highest_tests_df
```

**Q: Create a dataframe with 10 countires that have highest number of positive cases per million people.**

```python
positive_cases = merge_df.sort_values(by='cases_per_million' ,ascending = False).head(10)
positive_cases_df = pd.DataFrame(positive_cases).reset_index(drop =True)
positive_result_df =positive_cases_df[[
                    'location',
                    'continent', 
                    'population', 
                    'cases_per_million',
                    'hospital_beds_per_thousand',
                    'gdp_per_capita'
                    ]]
positive_result_df
```

**Q: Create a dataframe with 10 countires that have highest number of deaths cases per million people?**

```python
highest_number_of_deaths = merge_df.sort_values(by='deaths_per_million' ,ascending = False).head(10)
highest_number_of_deaths_df = pd.DataFrame(highest_number_of_deaths).reset_index(drop =True)
deaths_result_df =highest_number_of_deaths_df[[
                    'location',
                    'continent', 
                    'population', 
                    'deaths_per_million',
                    'hospital_beds_per_thousand',
                    'gdp_per_capita'
                    ]]
deaths_result_df
```

**Q: Count number of countries that feature in both the lists of "highest number of tests per million" and "highest number of cases per million".**

```python
tests_countries = highest_tests_df['location']
cases_countries = positive_result_df['location']

common_countries = set(tests_countries).intersection(set(cases_countries))
common_countries

num_common_countries = len(common_countries)
print(f"Number of countries that feature in both lists: {num_common_countries}")
print("Countries are: {}".format(common_countries))
```
